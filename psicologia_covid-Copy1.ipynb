```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./test_tables/covid19MH_20201027_geral_MLdata.csv', sep = ';')
```

```python
new_feat = []#list(data.columns)[42:141]
new_feat += ['BSI_23','BSI_36','BSI_20','BSI_17','BSI_38','BSI_46','BSI_31','BSI_10','BSI_53',
             'WHOQoL_6','WHOQoL_17','WHOQoL_18','WHOQoL_19','WHOQoL_20','WHOQoL_21','WHOQoL_13',
            'WHOQoL_23','WHOQoL_24','WHOQoL_5.2','WHOQoL_26', 'lost_familyfriend','work_homeoffice_now']
```

```python
result_cols = []
for col in data.columns:
    if 'whoqol' in col:
        result_cols.append(col)
    elif 'BSI_' in col and len(col)>6:
        result_cols.append(col)
```

```python
bsi = []
whoqol = []
for col in data.columns:
    if 'whoqol' in col:
        whoqol.append(col)
    elif 'BSI_' in col and len(col)>6:
        bsi.append(col)
```

```python
bsi_feat = []
whoqol_feat = []
for col in data.columns:
    if 'WHOQoL' in col:
        whoqol_feat.append(col)
    elif 'BSI_' in col and len(col)<=6:
        bsi_feat.append(col)
```

```python
data[bsi_feat].head()
```

```python

data[result_cols] = data[result_cols].replace(' ', np.nan)
data = data.replace(' ', np.nan)
```

```python
data.dropna(subset = result_cols, inplace = True)
```

```python
data.drop(['ID_covid', 'sample', 'date_created','consent', 'sgID', 'date_birth', 'region','city_br','Occupation_other', 'working_hours','religion_other', 'cceb_class',
'social_support_n_family','social_support_n_friends'],axis = 1, inplace = True)
```

```python
data[bsi] = data[bsi].apply(lambda x: x.str.replace(',','.'))
data[whoqol] = data[whoqol].apply(lambda x: x.str.replace(',','.'))
```

```python
data[bsi_feat] = data[bsi_feat].astype(float)

data[whoqol_feat] = data[whoqol_feat].astype(float)
```

```python
for col in data.columns:
    try:
        data[col] = data[col].astype(float)
    except:
        (print(col))
```

```python
import numpy as np # linear algebra
from numpy.random import seed
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from keras.layers import Input, merge
from keras.layers.core import  Dense, Dropout
from keras.models import Model
from tensorflow.python.client import device_lib
import keras
from keras import optimizers
from keras.layers.advanced_activations import LeakyReLU
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
data.fillna(-1, inplace = True)
Y1 = data['BSI_GSI']
Y2 = data[whoqol]
#data['mean'] = np.round(data['mean']/10)

#Y2 = data['"ARTERIAL_DIAS"']
data.drop(whoqol + bsi ,axis = 1, inplace = True)
X = data[bsi_feat]
X2 = data[new_feat]
```

# Train 1

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, Y1, test_size=0.33, random_state=42)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(
     X2, Y2, test_size=0.33, random_state=42)
```

```python
encoding_dim = 1
ncol = len(X2_train.columns)
keras.backend.clear_session()
```

```python
keras.backend.clear_session()
input_dim = Input(shape = (ncol, ))
# Encoder Layers

encoded1 = Dense(1000, activation = 'tanh')(input_dim)
drop1 = Dropout(0.4)(encoded1)
encoded2 = Dense(500, activation = 'tanh')(drop1)
encoded3 = Dense(200, activation = 'tanh')(encoded2)
encoded4 = Dense(150, activation = 'tanh')(encoded3)
encoded5 = Dense(80, activation = 'tanh')(encoded4)
decoded13 = Dense(1, activation = 'sigmoid')(encoded5)

# Combine Encoder and Deocder layers
autoencoder = Model(inputs = input_dim, outputs = [decoded13])

# Compile the Model
autoencoder.compile(optimizer = 'sgd', loss = 'mean_absolute_error')
```

```python
es = keras.callbacks.EarlyStopping(monitor='val_loss',
                              min_delta=0,
                              patience=1000,
                              verbose=0, mode='auto',
                                  restore_best_weights=True)




callbacks = [es]
autoencoder.fit(x = X2_train, y = y_train, epochs = 1500, 
                batch_size = 128, shuffle = True,
                 validation_data=(X2_test, y_test),
               callbacks = callbacks
                
               )
```
